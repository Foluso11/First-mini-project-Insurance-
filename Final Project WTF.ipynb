Use the data above, to answer the following questions:
- Relationship between BMI, charges and smoker
- Relationship between age and charges

```python
sns.scatterplot(x ='bmi', y ='charges', hue = 'smoker', data = insurance)
```

The above graph depicts the relationship between bmi, charges and smoker and from there, it can be deduced that there is a relationship between all three vaariables but it isn't a strong relationship almost no relationship.

```python
sns.lmplot(x ='age', y ='charges', data = insurance)
```

From this graph, there is not so strong a relationship between age and charges. This means age doesn't determine charges.
